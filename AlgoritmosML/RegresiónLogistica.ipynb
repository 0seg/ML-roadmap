# Regresión logistica

**Qué es la regresión logística?** La regresión logística es un algoritmo de clasificación que se utiliza para predecir la probabilidad de que una instancia pertenezca a una clase específica. A diferencia de la regresión lineal, que predice valores continuos, la regresión logística se utiliza para problemas de clasificación binaria o multiclase.

## Ejercicio sobre Regresión Logística

Este notebook contiene un ejercicio práctico sobre regresión logística. 

Usaremos un dataset llamado Fake_Real_News_Dataset.csv, que contiene noticias clasificadas como falsas o verdaderas. El objetivo es construir un modelo de regresión logística que pueda predecir si una noticia es falsa o verdadera basándose en ciertas características.

### 1. Lectura de datos

```python
import pandas as pd
```

```python
# Cargamos el dataset
df_fake = pd.read_csv('datasets/Fake_Real_News_Dataset/Fake.csv')
df_real = pd.read_csv('datasets/Fake_Real_News_Dataset/True.csv')
```

```python
# Hagamos una exploración rápida de los datos

df_fake.head()
```

```python
# Exploración de los datos reales

df_real.head()
```

```python
# Añadimos una columna "label" a cada dataframe para indicar si la noticia es falsa o verdadera

df_fake['label'] = 'False'
df_real['label'] = 'True'
```

```python
# Ahora unimos ambos dataframes en uno solo

# ignore_index=True asegura que los índices se reenumeren en el dataframe combinado
df = pd.concat([df_real, df_fake], ignore_index=True) 
```

```python
# Comprobamos que los datos se han cargado correctamente y que la unión de los dataframes ha sido exitosa

print(f"Noticias reales: {len(df_real)}")
print(f"Noticias falsas: {len(df_fake)}")
print(f"Total de noticias: {len(df)}")
```

```python
# Usamos df.head() para ver las primeras filas del dataframe combinado y asegurarnos de que todo está correcto
df.head()
```

```python
# Usamos df.tail() para ver las últimas filas del dataframe combinado y asegurarnos de que todo está correcto
df.tail() 
```

### 3. Preprocesamiento de datos

Aquí nos encontramos con uno de los grandes retos del aprendizaje automático: la preparación de los datos. Esta etapa es crucial para garantizar que el modelo pueda aprender de manera efectiva. Incluye la limpieza de datos, la transformación de variables y la creación de nuevas características.

Las noticias pueden contener: 

- **Código HTML:** (etiquetas como `<p>`, `<a>`, `<div>`, etc.)
- **URLs:** enlaces a otros sitios web.
- **Simbolos de puntuación:** como comas, puntos, signos de interrogación, etc.
- **Palabras comunes:** que aparecen en todas las noticias y no ayudan a distinguir entre noticias falsas y verdaderas.(por ejemplo: "the", "is", "and", etc.)

Antes de entrenar el modelo, es importante limpiar y preprocesar los datos para eliminar estos elementos que no aportan valor al modelo. Esto puede incluir la eliminación de etiquetas HTML, la eliminación de URLs, la eliminación de símbolos de puntuación y la eliminación de palabras comunes.
