# Introducción a Matplotlib

Matplotlib es una biblioteca de visualización de datos en Python que permite crear gráficos estáticos, animados e interactivos. Es ampliamente utilizada en la comunidad científica y de análisis de datos debido a su flexibilidad y capacidad para generar gráficos de alta calidad.

- **Permite crear una amplia variedad de gráficos**: desde gráficos de líneas y barras hasta histogramas y gráficos de dispersión.

# Import

```python
import matplotlib.pyplot as plt
```

# Representación de datos con Matplotlib

**Punto clave**: Si a una funciona de trazado de gráficos le pasamos un array de datos, Matplotlib lo interpreta como el eje Y y genera automáticamente un array de índices para el eje X.

```python
lista = [1, 2, 3, 4, 5, 3, 5, 2, 20]

plt.plot(lista)
plt.title("Gráfico de ejemplo")
plt.show()
```

También podemos especificar los valores del eje X pasando un segundo array de datos a la función de trazado.

```python
lista2 = [10, 20, 30, 40, 50, 60, 70, 80, 90]
plt.plot(lista2, lista)
plt.title("Gráfico con eje X especificado")
plt.show()
```

También podemos modificar las longitudes de los ejes, para que la figura se vea más proporcionada.

```python
plt.plot(lista2, lista)
plt.axis([0, 100, 0, 25])  # Modifica los límites de los ejes: [xmin, xmax, ymin, ymax]
plt.title("Gráfico con ejes modificados")
plt.show()
```

Graficando una función matemática, podemos utilizar la función `numpy.linspace` para generar un conjunto de valores equidistantes en un intervalo específico. Esto nos permite crear gráficos suaves y continuos.

```python
import numpy as np

x = np.linspace(-10, 10, 1000)  # Genera 1000 valores equidistantes entre -10 y 10
y = np.sin(x)  # Calcula el seno de cada valor de x

plt.plot(x, y)
plt.title("Gráfico de la función seno")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.grid()  # Agrega una cuadrícula al gráfico
plt.show()
```

Podemos superponer múltiples gráficos en la misma figura utilizando la función `plt.plot()` varias veces antes de llamar a `plt.show()`. Esto es útil para comparar diferentes conjuntos de datos en un solo gráfico.

```python
import numpy as np

x = np.linspace(-10, 10, 100)  # Genera 1000 valores equidistantes entre -10 y 10

y1 = np.sin(x)  # Calcula el seno de cada valor de x
y2 = np.cos(x)  # Calcula el coseno de cada valor de x

plt.plot(x, y1, x, y2)  # Superpone los gráficos de seno y coseno
plt.title("Gráfico de las funciones seno y coseno")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.legend(["Seno", "Coseno"])  # Agrega una leyenda para identificar cada función
plt.grid()  # Agrega una cuadrícula al gráfico
plt.show()


```

También podemos crear dos graficos que no se superpongan. Utilizando la función `plt.subplot()`, podemos dividir la figura en subgráficos y trazar diferentes gráficos en cada uno de ellos. Esto es útil para comparar visualmente diferentes conjuntos de datos sin que se superpongan.

```python
import numpy as np

x = np.linspace(-10, 10, 100)  # Genera 1000 valores equidistantes entre -10 y 10

y1 = np.sin(x)  # Calcula el seno de cada valor de x
y2 = np.cos(x)  # Calcula el coseno de cada valor de x

plt.subplot(1, 2, 1)  # Crea un subgráfico en la primera fila
plt.plot(x, y1)  # Dibuja el gráfico de seno


plt.subplot(1, 2, 2)  # Crea un subgráfico en la segunda fila
plt.plot(x, y2)  # Dibuja el gráfico de coseno

plt.title("Gráfico de las funciones seno y coseno")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.legend(["Seno", "Coseno"])  # Agrega una leyenda para identificar cada función
plt.grid()  # Agrega una cuadrícula al gráfico
plt.show()
```

También podemos especificar el tamaño de la figura utilizando la función `plt.figure(figsize=(ancho, alto))`. Esto nos permite ajustar el tamaño del gráfico según nuestras necesidades.

```python
import numpy as np

plt.figure(figsize=(12, 6))  # Ajusta el tamaño de la figura a 12x6 pulgadas

x = np.linspace(-10, 10, 100)  # Genera 1000 valores equidistantes entre -10 y 10

y1 = np.sin(x)  # Calcula el seno de cada valor de x
y2 = np.cos(x)  # Calcula el coseno de cada valor de x

plt.subplot(1, 2, 1)  # Crea un subgráfico en la primera fila
plt.plot(x, y1)  # Dibuja el gráfico de seno
plt.grid()  # Agrega una cuadrícula al gráfico

plt.subplot(1, 2, 2)  # Crea un subgráfico en la segunda fila
plt.plot(x, y2)  # Dibuja el gráfico de coseno

plt.title("Gráfico de las funciones seno y coseno")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.legend(["Seno", "Coseno"])  # Agrega una leyenda para identificar cada función
plt.grid()  # Agrega una cuadrícula al gráfico
plt.show()
```

# Scatter plots

**¿Qué es un scatter plot?** Un scatter plot (o diagrama de dispersión) es un tipo de gráfico que utiliza puntos para representar la relación entre dos variables. Cada punto en el gráfico representa un par de valores (x, y), lo que permite visualizar patrones, tendencias y posibles correlaciones entre las variables.

```python
# Generando datos aleatorios para el scatter plot

from numpy.random import randn

x = randn(100)  # Genera 100 valores aleatorios con distribución normal
y = randn(100)  # Genera 100 valores aleatorios con distribución normal
```

```python
plt.scatter(x, y)  # Crea un scatter plot con los datos generados
plt.title("Diagrama de dispersión (Scatter Plot)")
plt.xlabel("Variable X")
plt.ylabel("Variable Y")
plt.show()  # Muestra el gráfico
```

También podemos superponer múltiples scatter plots en la misma figura utilizando la función `plt.scatter()` varias veces antes de llamar a `plt.show()`. Esto es útil para comparar diferentes conjuntos de datos en un solo gráfico.

```python
from numpy.random import randn

x, y = randn(2, 100)  # Genera 100 pares de valores aleatorios con distribución normal
x2, y2 = randn(2, 100)  # Genera otros 100 pares de valores aleatorios con distribución normal

plt.scatter(x, y, color='blue', label='Conjunto 1')  # Primer conjunto de datos
plt.scatter(x2, y2, color='red', label='Conjunto 2')
plt.title("Diagrama de dispersión con múltiples conjuntos de datos")
plt.legend()  # Agrega una leyenda para identificar cada conjunto de datos
plt.show()  # Muestra el gráfico
```

# Histogramas

**¿Qué es un histograma?** Un histograma es un tipo de gráfico que representa la distribución de un conjunto de datos dividiéndolo en intervalos (bins) y mostrando la frecuencia de los datos en cada intervalo. Es útil para analizar la forma, dispersión y tendencia central de los datos.

```python
data = [randn(1000)]  # Genera 1000 valores aleatorios con distribución normal

plt.subplot(2, 1, 1) 

plt.hist(data, bins=10, rwidth=0.8) 

# rwidth ajusta el ancho de las barras del histograma. 
# Bins especifica el número de intervalos en los que se divide el rango de datos.

plt.title("Histograma de los conjuntos de datos")
plt.xlabel("Intervalos (bins)")
plt.ylabel("Frecuencia")

plt.show()  
```

# Guardar los gráficos

También podemos guardar las figuras generadas en diferentes formatos de archivo utilizando la función `plt.savefig()`. Esto nos permite exportar los gráficos para su uso en presentaciones, informes o publicaciones.

```python
import numpy as np

x = np.linspace(-10, 10, 1000)  # Genera 1000 valores equidistantes entre -10 y 10
y = np.sin(x)  # Calcula el seno de cada valor de x

plt.plot(x, y)
plt.title("Gráfico de la función seno")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.grid()  # Agrega una cuadrícula al gráfico
plt.savefig("grafico_seno.png", transparent=True)  # Guarda el gráfico en un archivo PNG

#transparent=True hace que el fondo del gráfico sea transparente, lo que puede ser útil para superponer el gráfico en otros documentos o presentaciones.
```
