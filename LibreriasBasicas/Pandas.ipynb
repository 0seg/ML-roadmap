# Introducción a Pandas

**Pandas** es una biblioteca de Python que proporciona estructuras de datos y herramientas de análisis de datos de alto rendimiento y fáciles de usar. Es especialmente útil para trabajar con datos tabulares, como hojas de cálculo o bases de datos.


**Ventajas de Pandas**:
  - Facilita la manipulación y análisis de datos.
  - Permite trabajar con datos faltantes y realizar operaciones de limpieza.
  - Ofrece herramientas para combinar, fusionar y agrupar datos.
  - Es compatible con otras bibliotecas de Python, como NumPy y Matplotlib, lo que permite realizar análisis más avanzados y visualizaciones.



# Import 

```python
import pandas as pd
```

# Estructuras de datos principales de Pandas

Las estructuras de datos principales de Pandas son **Series** y **DataFrame**, **Panel**. A continuación, se describen brevemente:

- **Series**: Una Serie es una estructura de datos unidimensional que puede contener cualquier tipo de datos (enteros, cadenas, flotantes, etc.). Cada elemento de la Serie tiene un índice asociado, lo que permite acceder a los datos de manera eficiente.

- **DataFrame**: Un DataFrame es una estructura de datos bidimensional que puede contener datos de diferentes tipos (enteros, cadenas, flotantes, etc.). Se puede pensar en él como una tabla de datos o una hoja de cálculo.

- **Panel**: Un Panel es una estructura de datos tridimensional que puede contener múltiples DataFrames. Es menos utilizado actualmente, ya que se pueden lograr similares resultados con otras estructuras de datos más modernas.  

```python
# Creación de una Serie

s = pd.Series([1, 2, 3, 4, 5])
s
```

```python
# Creación de una Serie inicializandolo con un diccionario de Python

data = {'a': 10, 'b': 20, 'c': 30}
s2 = pd.Series(data)
s2
```

```python
# Creación de una Serie inicializandolo con algunos de los elementos del diccionario de Python

data = {'a': 10, 'b': 20, 'c': 30}
s3 = pd.Series(data, index=['a', 'c'])
s3
```

```python
# Creación de una Serie inicializandolo con un escalar

s = pd.Series(5, index=['a', 'b', 'c'])
s
```

# Acceso a los elementos de una Serie

Cada elemento en una Serie tiene un identificador único que se denomina index label. El index label se puede usar para acceder a los elementos de la Serie de manera eficiente. A continuación, se muestran algunos ejemplos de cómo acceder a los elementos de una Serie:


```python
# Creación de una Serie

s = pd.Series([1, 2, 3, 4, 5], index=['num1', 'num2', 'num3', 'num4', 'num5'])
```

```python
# Accediendo al tercer elemento de la Serie utilizando su index label.

s['num3']
```

```python
# Loc es el metodo estandar para acceder a un elemento por atributo.

s.loc['num3']
```

```python
# Iloc es el metodo estandar para acceder a un elemento por posicion.

s.iloc[2]
```

# Operaciones aritméticas con Series


```python
# Creación de una Serie

s = pd.Series([2, 4, 6, 8, 10])
s
```

```python
# Las Series son similares y compatibles con los arrays de Numpy, por lo que podemos realizar operaciones aritméticas con ellas.

import numpy as np

# Ufunc de Numpy para sumar 5 a cada elemento de la Serie.

np.add(s, 5)
```

```python
# Elevar a la potencia de 2 cada elemento de la Serie.

np.power(s, 2)
```

```python
# Raiz cuadrada de cada elemento de la Serie.

np.sqrt(s)
```

# Representación gráfica de Series

```python
# Creación de una Serie para temperaturas

temperaturas = [4.0, 5.5, 6.0, 7.2, 8.1, 9.3, 10.5, 6.7, 5.8, 4.9, 3.2, 2.1]
s_temperaturas = pd.Series(temperaturas, name='Temperaturas')
s_temperaturas
```

```python
# Representación gráfica de la Serie de temperaturas utilizando matplotlib.

%matplotlib inline
import matplotlib.pyplot as plt

s_temperaturas.plot()
plt.show()
```

# Creación de un DataFrame


```python
# Creación de un DataFrame inicializado con un diccionario que contiene series de Pandas.

personas = {
    "peso": pd.Series([70, 80, 90, 65], index=['Juan', 'Ana', 'Luis', 'Marta']),
    "altura": pd.Series([1.75, 1.65, 1.80, 1.60], index=['Juan', 'Ana', 'Luis', 'Marta']),
    "hijos": pd.Series([2, 1, 3, 0], index=['Juan', 'Ana', 'Luis', 'Marta'])

}

df = pd.DataFrame(personas)
df
```

```python
# Creación de un DataFrame inicializandolo con una lista de listas de Python.

valores = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]

df = pd.DataFrame(valores, columns=['A', 'B', 'C'], index=['Fila1', 'Fila2', 'Fila3'])

df
```
